```python
import MySQLdb
import sqlalchemy
```

```python
user = '**' ## adicione o username
password = '**' ## adicione password
host = 'localhost'
database = '**' ## nome de sua tabela no sql

string_conexao = f'mysql://{user}:{password}@{host}/{database}'
string_conexao
```

```python
engine = sqlalchemy.create_engine(string_conexao)
conn = engine.connect()
```

```python
db_jogador = lista_csv[['jogador', 'genero', 'data_nascimento','num_vitorias', 'num_derrotas', 'total_partidas', 'pais', 'jogo']].drop_duplicates()
jogo = list(lista_csv.jogo.unique())
pais = list(lista_csv.pais.unique())
```

```python
db_jogador.rename(columns={'jogador':'nome_jogador', 'pais':'cod_pais', 'jogo':'cod_jogo'} , inplace=True)
```

```python
## mudar disposição da data para inserir no SQL
db_jogador['data_nascimento'] = pd.to_datetime(db_jogador['data_nascimento'], format= '%d/%m/%Y')
db_jogador.head()
```

```python
# inserindo dados de país no SQL
for paises in pais:
    try:
        query = f'''insert into pais (nome_pais)
                    values('{paises}')'''
        conn.execute(query)
        print(f'Registro inserido com sucesso: {paises}')

    except Exception as e:
        print(f'Não foi possível inserir o registro {paises}. O erro encontrado foi: {e}')
```

```python
sql_pais = pd.read_sql('pais', con=conn)
```

```python
# codificando paises
dict_pais = sql_pais.set_index(sql_pais.nome_pais)['cod_pais'].to_dict()
dict_pais
```

```python
# inserindo informações de jogo 
for jogos in jogo:
    try:
        query = f'''insert into jogo (nome_jogo)
                    values('{jogos}')'''
        conn.execute(query)
        print(f'Registro inserido com sucesso: {jogos}')

    except Exception as e:
        print(f'Não foi possível inserir o registro {jogos}. O erro encontrado foi: {e}')
```

```python
# convertendo a lista jogos em dicionário pandas
sql_jogo = pd.read_sql('jogo', con=conn)
sql_jogo
```

```python
# codificando jogos
dict_jogo = sql_jogo.set_index(sql_jogo.nome_jogo)['cod_jogo'].to_dict()
dict_jogo
```

```python
# verificando o dataframe
db_jogador.head(5)
```

```python
# passando a codificação de pais e jogo para o dataframe
db_jogador.cod_pais = db_jogador.cod_pais.map(dict_pais) 
db_jogador.cod_jogo = db_jogador.cod_jogo.map(dict_jogo) 
db_jogador.head()
```

```python
# inserindo dados do dataframe no SQL 
for jogador in db_jogador.itertuples():
    nome_jogador = jogador.nome_jogador
    genero = jogador.genero
    data_nascimento = jogador.data_nascimento
    num_vitorias = jogador.num_vitorias
    num_derrotas = jogador.num_derrotas
    total_partidas = jogador.total_partidas
    cod_pais = jogador.cod_pais
    cod_jogo = jogador.cod_jogo
    


    try:
        query = f'''insert into jogador (nome_jogador, genero, data_nascimento, num_vitorias, num_derrotas, total_partidas, cod_pais, cod_jogo)
                    values('{nome_jogador}', '{genero}', '{data_nascimento}', '{num_vitorias}', '{num_derrotas}', '{total_partidas}', '{cod_pais}', '{cod_jogo}')'''
        conn.execute(query)
        print(f'Registro inserido com sucesso: {nome_jogador}')

    except Exception as e:
        print(f'Não foi possível inserir o registro {nome_jogador}. O erro encontrado foi: {e}')
```
